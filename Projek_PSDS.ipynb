Komoditas apa yang termasuk tertinggi dalam hasil budidaya

```python
import matplotlib.pyplot as plt 
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('jumlahproduksibenih.csv')
df
```

```python
hasil_budidaya = pd.DataFrame(df[['Benih','Budidaya','Jumlah']])
hasil_budidaya
```

```python
benih_pertahun = hasil_budidaya.groupby("Benih").sum()
benih_pertahun
```

```python
komoditas_terbanyak = benih_pertahun.nlargest(1,'Jumlah')
komoditas_terbanyak
```

```python
#Visualisasi Data 
plt.clf()
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(data=df, x='Benih', height='Jumlah')
plt.xticks(rotation='vertical')
plt.show()
```

3 daerah yang termasuk produsen tertinggi 

```python
import matplotlib.pyplot as plt 
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('produksibudidayanasional.csv')
df
```

```python
produsen = pd.DataFrame(df[['NamaProvinsi','Tahun','Volume']])
produsen
```

```python
produsen_pertahun = produsen.groupby("NamaProvinsi").sum()
produsen_pertahun
```

```python
produsen_pertahun.nlargest(3,'Volume')
```

```python
#Visualisasi Data 
plt.clf()
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(data=df, x='NamaProvinsi', height='Volume')
plt.xticks(rotation='vertical')
plt.show()
```
